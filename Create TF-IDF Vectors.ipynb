```python
i=0
def Convert(string): 
    li = list(string.split(" ")) 
    return li 

def listToString(a):  
    
    # initialize an empty string 
    str1 = " " 
    
    # return string   
    return (str1.join(a)) 

# Driver code     
str1 = "deneme deneme"
print(Convert(str1)) 
dokumanla = []
d_dokumanla=[]
uzunluk=len(haber)
while i < uzunluk:
    dokumanla.append(haber[i]['Document'])
    print(i)
    i += 1
            
d_dokumanla=listToString(dokumanla)
d_dokumanla=d_dokumanla.replace("\"","")
d_dokumanla=d_dokumanla.replace("\'","")
d_dokumanla=d_dokumanla.replace("-"," ")
d_dokumanla=d_dokumanla.replace("\n"," ")
uzunluk=len(d_dokumanla)
i=0
words= []

print(10)
# create the transform
vectorizer = TfidfVectorizer(max_features=1200)
print(20)
# tokenize and build vocab
vectorizer.fit(Convert(d_dokumanla))
print(30)
# summarize
y=0
y="Başbakan Erdoğan ocak ayı için 500 liranın altında maaş alan emeklilere 60 liralik zam bin 500 liranın üzerinde"
print('vocabulary: ', vectorizer.vocabulary_)
vector=vectorizer.transform([y])
print('\nvectors: ', vector.toarray())
```
